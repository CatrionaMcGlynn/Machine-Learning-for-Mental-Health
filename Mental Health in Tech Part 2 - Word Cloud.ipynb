This workbook displays a word cloud for the most popular words in the comments field.

```
import pandas as pd

import matplotlib.pyplot as plt

import re
from collections import Counter

from wordcloud import WordCloud

import nltk 
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize
```

```
df = pd.read_csv("survey.csv")
df = df[['comments']]
df = df.dropna()
```

```
nltk.download('stopwords')
nltk.download('punkt')
```

```
def regex_clean(txt, regex):
    """Replace any text matching the regex

    Parameters
    ----------
    txt : string
        A text string that you want to parse and remove matches
    regex : string 
        A text string of the regex pattern you want to match

    Returns
    -------
    The same txt string with the matches removes
    """

    return " ".join(re.sub(regex, "", txt).split())
```

```
df['comments'] = df['comments'].astype(str)
```

```
df['comments'] = df['comments'].apply(str.lower)
df['tokens'] = df['comments'].apply(word_tokenize)
```

```
stpwrd = nltk.corpus.stopwords.words('english')

punc = '!"#$%&()*+, -./:;<=>?@[\]^_`{|}~”“\''

punc = [x for x in punc]
```

```
stpwrd.extend(["n't", "'m", "'s", "'ve", 'also', 'would', 'could', 'much',
               'might', 'many', 'may'])
stpwrd.extend(punc)
```

```
df['tokens'] = df['tokens'].apply(lambda x:[words for words in x if words not in stpwrd])

token_counts = Counter()
df['tokens'].apply(token_counts.update)
#token_counts
```

```
wc = WordCloud(width=400, height=200, max_words=50, colormap="Reds" ).generate_from_frequencies(token_counts)
plt.figure(figsize=(24,20))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.savefig("MHwordcloud.png")
plt.show()
```
